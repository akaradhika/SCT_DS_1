### Step 2–7: Import Libraries, Analyze Data & Visualize

In this section, we:
- Import the required libraries (`pandas` for data handling and `matplotlib` for visualization)
- Load the uploaded CSV file containing 2024 population data
- Count how many countries are in the dataset
- Sort and display the **top 20 most populous countries**
- Create a **bar chart** to visualize these top 20 countries
- Plot a **histogram** to show the distribution of populations across all countries


```python
# STEP 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# STEP 3: Load the uploaded file (replace filename if different)
df = pd.read_csv('population_2024.csv')

# STEP 4: Check the number of countries
print(f"Total countries in the dataset: {df.shape[0]}")

# STEP 5: Show top 20 most populous countries
top20 = df.sort_values(by='Population (Thousands)', ascending=False).head(20)
print("Top 20 countries by population in 2024:")
print(top20[['Country', 'Population (Thousands)']])

# STEP 6: Bar chart of top 20 countries
plt.figure(figsize=(14, 7))
plt.bar(top20['Country'], top20['Population (Thousands)'], color='skyblue')
plt.title('Top 20 Most Populous Countries in 2024')
plt.xlabel('Country')
plt.ylabel('Population (Thousands)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# STEP 7: Histogram of population distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Population (Thousands)'], bins=15, color='orange', edgecolor='black')
plt.title('Population Distribution of Countries (2024)')
plt.xlabel('Population (Thousands)')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
